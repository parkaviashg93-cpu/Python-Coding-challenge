```python
# Student Grades & Hobbies Tracker

# Step 1: Create a dictionary with student names as keys and grades as values
student_grades = {"Ashwini": "A","Dhariya": "B+","Dhana": "C"}
print("Initial Student Grades:")
print(student_grades)
```

```python
# Step 2: Allow the user to add or update grades for at least 3 students
for i in range(3):
    name = input("\nEnter student name: ")
    grade = input("Enter grade (e.g., A, B+, C): ")
    student_grades[name] = grade
    print(f"Updated {name}'s grade to {grade}")
print("\nFinal Student Grades:")
print(student_grades)
```

```python
# Step 3: Create a set of hobbies and allow the user to input at least 3 hobbies
hobbies = set()
for i in range(3):
    hobby = input("\nEnter a hobby: ")
    hobbies.add(hobby)
print("\nHobbies Set:")
print(hobbies)
```

```python
# Step 4: Convert the set of hobbies into a frozenset and display it
frozen_hobbies = frozenset(hobbies)

print("\nFrozen Hobbies (Immutable):")
print(frozen_hobbies)
```
