```python
#Topic:Operators, Conditional Statements
#Project Title: BMI Calculator
#Scenario:
 A health organization wants a tool to help users calculate their Body Mass Index (BMI) and understand their health category based on it.
```

```python
#Tasks:
#Ask the user to enter their weight (in kilograms) and height (in meters).
weight=int(input("Enter the weight in kilograms:"))
height=float(input("Enter the height in meters:"))
print(weight)
print(height)
```

```python
#Calculate their BMI using the formula BMI = weight / (height × height).
BMI = weight / (height * height)
print("\nYour BMI is:", round(BMI, 2))
```

```python
#Classify BMI using conditional statements
if BMI <18.5:
    print("Health Status: Underwieght")
elif BMI >=18.5 and BMI < 25:
    print("Health Status: Normal")
elif BMI >=25 and BMI < 30:
    print("Health Status: Overweight")
else:
    print("Health Status: Obese")
```
