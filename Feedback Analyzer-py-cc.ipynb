```python
#project title: Feedback Analyzer
# Step 1: Ask for user input
feedback=input("Enter Your Feedback: ")
```

```python
# Step 2: Count how many words are in the sentence
word_count = len(feedback.split())
print("Total number of words:", word_count)
```

```python
# Step 3: Check how many times 'data' appears (case-insensitive)
data_count=feedback.lower().count("data")
print("The word 'data' appears:", data_count, "times")
```

```python
# Step 4: Extract first 5 and last 5 characters
first_five = feedback[:5]
last_five = feedback[-5:]
print("First 5 characters:", first_five)
print("Last 5 characters:", last_five)
```

```python
# Step 5: Reverse the sentence
reversed_feedback=feedback[::-1]
print("Reversed feedback:", reversed_feedback)
```
