```python
# Step 1: Create a 2D list (each row = [Name, Phone, Email])
contacts = [
    ["Arya", "9934746243", "arya43@gmail.com"],
    ["Banu", "9237574652", "banu4652@gmail.com"],
    ["Chitra", "9675342535", "chita2535@gmail.com"]
]
print(contacts)
```

```python
# Step 2: Allow user to add a new contact
new_contact = ["Deva", "9736548165", "deva8165@gmail.com"]
contacts.append(new_contact)

print("After adding new contact:")
print(contacts)
```

```python
# Step 3: Remove the last item
contacts.pop()
print("\nAfter removing last contact:")
print(contacts)
```

```python
# Step 4: Get new items from user input
name=input("Enter Contact Nmae:")
phone=input("Enter Phone Number:")
email=input("Enter Email Address:")
contacts.append([name,phone,email])
print("\nafter adding user input contact:")
print(contacts)
```

```python
# Step 5: Sort the list in ascending order (by name here)
contacts.sort(key=lambda x: x[0])
print("\nSorted contacts by name:")
print(contacts)
```
