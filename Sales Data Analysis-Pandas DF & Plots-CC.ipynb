```python
#Basic Cleaning
#Check Missing Values
df.isnull().sum()
```

```python
#Remove duplicate rows
df.drop_duplicates(inplace=True)
df
```

```python
#matplot
import matplotlib.pyplot as plt
```

```python
#top 5 city total sales
import matplotlib.pyplot as plt
city_sales = df.groupby("City")["Sales_Amount"].sum().sort_values(ascending=False).head(5)

plt.figure()
city_sales.plot(kind="bar")
plt.title("Top 5 Cities by Total Sales")
plt.xlabel("City")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Monthly Sales Trend (Line Chart)
df["Month"] = df["Order_Date"].dt.month
monthly_sales=df.groupby("Month")["Sales_Amount"].sum()
plt.figure()
monthly_sales.plot(kind="line", color="red")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Sales vs Profit (Scatter Plot)
plt.figure()
plt.scatter(df["Sales_Amount"], df["Profit"], color="orange")
plt.title("Sales vs Profit")
plt.xlabel("Sales Amount")
plt.ylabel("Profit")
plt.show()
```

```python
# Pie Chart – Sales by Segment
segment_sales = df.groupby("Segment")["Sales_Amount"].sum()

plt.figure(figsize=(3,3))
plt.pie(segment_sales,
        labels=segment_sales.index,
        autopct='%1.1f%%',
        startangle=90)

plt.title("Sales Distribution by Segment")
plt.show()
```

```python
#Histogram Distribution

plt.figure(figsize=(8,5))
plt.hist(df["Sales_Amount"], bins=30, color="skyblue", edgecolor="black")
plt.title("Sales Amount Distribution")
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
#Box Plot (Outliers)
plt.figure(figsize=(7,5))
plt.boxplot(df["Profit"])
plt.title("Profit Distribution with Outliers")
plt.ylabel("Profit")
plt.show()
```

```python
#Seaborn
import seaborn as sns
```

```python
#Heatmap (Correlation)
plt.figure(figsize=(8,6))
corr = df[["Sales_Amount","Profit","Units_Sold","Unit_Price"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Pair Plot (Relationships)
sns.pairplot(df[["Sales_Amount","Profit","Units_Sold","Unit_Price"]])
plt.show()
```

```python
#Violin Plot (Distribution + Density)
plt.figure(figsize=(8,5))
sns.violinplot(data=df, x="Segment", y="Profit")
plt.title("Profit Distribution by Segment (Violin Plot)")
plt.show()
```

```python
#Swarm Plot (Actual data points)
plt.figure(figsize=(8,5))
sns.swarmplot(data=df, x="Segment", y="Profit")
plt.title("Profit Distribution by Segment (Swarm Plot)")
plt.show()
```

```python
#Density Plot (KDE Plot)
plt.figure(figsize=(7,5))
sns.kdeplot(data=df, x="Sales_Amount", fill=True)
plt.title("Sales Amount Density Plot")
plt.show()
```

# Project Title:Sales Data Analysis Project

This project analyzes a sales transaction dataset using Python libraries
such as Pandas, Matplotlib, Seaborn, and Plotly to identify trends,
distributions, correlations, and business insights.


## Dataset Description

The dataset contains 2500 sales transaction records with information
about customers, products, pricing, discounts, and profit.

Columns in the dataset:

- Order_ID – Unique identifier for each order
- Customer_Name – Name of the customer who placed the order
- Product_Category – Category of the product (e.g., Electronics, Furniture)
- Product_Name – Specific product sold
- Units_Sold – Number of units sold in the transaction
- Unit_Price – Price per unit of the product
- Discount_% – Percentage discount applied to the product
- Sales_Amount – Total sales value after applying discount
- Order_Date – Date when the order was placed
- City – City where the order was made
- Segment – Customer segment (Consumer, Corporate, Home Office)
- Profit – Profit generated from the order

## Data Cleaning Steps

The dataset was cleaned before analysis to ensure data quality.

Steps performed:

1. Checked duplicate rows using:
   df.duplicated().sum()

2. Result showed 0 duplicate rows, so no rows were removed.

3. Checked missing values using:
   df.isnull().sum()

4. Missing values were handled using:
   df.fillna(0)

5. The Order_Date column was converted to datetime format using:
   pd.to_datetime(df["Order_Date"], format="mixed", dayfirst=True)

6. A new column called Month was created for trend analysis:
   df["Month"] = pd.to_datetime(df["Order_Date"], format="mixed", dayfirst=True).dt.month
