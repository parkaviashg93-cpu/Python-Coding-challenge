```python
# Shopping Cart Billing System

subtotal = 0.0
item_count = 0

print("Welcome to the Shopping Cart Billing System")
print("Enter the price of each item.")
print("Enter 0 when you are done.\n")
```

```python
while True:
    price = float(input("Enter item price: "))

# Stop condition (sentinel value)
    if price == 0:
        print("\nBilling completed...")
        break

# Negative price check
    if price < 0:
        print("Invalid price! Please enter a positive value.\n")
        continue

# Valid price
    item_count += 1
    subtotal += price
    print(f"Item added. Current subtotal: ₹{subtotal}\n")

else:
    # This will run only if while loop ends normally (without break)
    print("Loop ended normally")
```

```python
# Discount calculation
if subtotal >= 1000:
    discount_rate = 0.20
elif subtotal >= 500:
    discount_rate = 0.10
elif subtotal >= 200:
    discount_rate = 0.05
else:
    discount_rate = 0.0

discount_amount = subtotal * discount_rate
final_total = subtotal - discount_amount

# Bill Summary
print("\n----- BILL SUMMARY -----")
print(f"Total items purchased : {item_count}")
print(f"Subtotal              : ₹{subtotal:.2f}")
print(f"Discount              : ₹{discount_amount:.2f}")
print(f"Final Total to Pay    : ₹{final_total:.2f}")
print("------------------------")
print("Thank you for shopping with us!")
```
