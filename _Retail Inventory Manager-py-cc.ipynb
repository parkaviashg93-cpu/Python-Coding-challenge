```python
#Project: Retail Inventory Manager
# Step 1: Create a tuple with fixed product categories
product_categories=("Electronics", "Stationery", "Clothing")
```

```python
# Step 2: Create a list of products that can be updated dynamically
products = ["Laptop", "Pencil", "Shirt"]
```

```python
# Step 3: Ask the user to enter a new product and add it to the list
new_product = input("Enter a new product to add: ")
products.append(new_product)
```

```python
# Step 4: Ask the user to input the index of a product to remove
print("Current products:", products)
remove_index = int(input("Enter the index of the product to remove: "))
# Check if index is valid
if 0 <= remove_index < len(products):
    removed_item = products.pop(remove_index)
    print(f"Removed product: {removed_item}")
else:
    print("Invalid index. No product removed.")
```

```python
# Step 5: Display the final list of products and product categories
print("\nFinal Product Categories (Tuple - Immutable):", product_categories)
print("Final Product List (List - Mutable):", products)
```
